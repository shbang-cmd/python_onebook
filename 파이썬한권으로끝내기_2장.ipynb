```python
# 선택한 값 변경하는 법
score = pd.DataFrame({'국어':[100, 80], '수학':[75, 90],
                      '영어':[90,95]}, index=['장화', '홍련'])
score
```

```python
score.loc['홍련','영어'] = 100
score  # 영어성적이 100으로 바뀌었다.
```

```python
score['국어'] = score['국어'] - 5
score  # 모든 학생의 국어점수가 5점 내려감
```

```python
# 행과 열 추가
# 행추가는 append()를 사용
new_student = pd.DataFrame({'국어':[70,85], '수학':[65,100],
                            '영어':[95,65]}, index=['콩쥐', '팥쥐'])
new_student
```

```python
#score =  score.append(new_student)
score = pd.concat([score, new_student], ignore_index=False)
score
```

```python
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
new_student1
```

```python
score = pd.concat([score, new_student1.to_frame().T])
score
```

```python
# Dictionary 자료형을 사용하여 행을 추가할 경우 인덱스로 사용할 정보가 없는 경우 ignore_index인자를 True로 설정하지 않으면 오류가 발생
# ignore_index를 True로 설정하면 기존인덱스가 전부 사라지기 때문에 주의하여 사용하여야 한다.
# DataFrame에는 DataFrame 타입의 자료구조끼리만 합칠 수 있다.
new_student2 = {'국어':75, '수학':80, '영어':80}
score = pd.concat([score, new_student2], ignore_index = False)  # Error : 딕셔너리를 합치려고 하면 안됨
```

```python
# DataFrame끼리만 연결되고 dict 타입은 연결이 안되므로 아래와 같이 한다.
new_student2_df = pd.DataFrame([new_student2], index=['달님'])
score = pd.concat([score, new_student2_df])
score
```

```python
# column 추가
science = [80, 70, 90, 85, 75, 70]
score['과학'] = science
score['학년'] = 1
score
```

```python
score.info()
```

```python
score.describe()
```

```python
# 기존 데이터프레임의 행의 수와 같다면 계산식의 결과를 새로운 컬럼으로 추가 가능
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

```python
# 행과 열의 삭제, 행과 열을 동시에 삭제는 불가능
score.drop('장화', inplace=True)  # 장화의 행 삭제
score
```

```python
score.drop(columns = ['과학', '학년', '총점'], inplace=True)
score
```
