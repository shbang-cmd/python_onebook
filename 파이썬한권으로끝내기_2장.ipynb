```python
# 선택한 값 변경하는 법
score = pd.DataFrame({'국어':[100, 80], '수학':[75, 90],
                      '영어':[90,95]}, index=['장화', '홍련'])
score
```

```python
score.loc['홍련','영어'] = 100
score  # 영어성적이 100으로 바뀌었다.
```

```python
score['국어'] = score['국어'] - 5
score  # 모든 학생의 국어점수가 5점 내려감
```

```python
# 행과 열 추가
# 행추가는 append()를 사용했는데 이제는 pd.concat() 이용
new_student = pd.DataFrame({'국어':[70,85], '수학':[65,100],
                            '영어':[95,65]}, index=['콩쥐', '팥쥐'])
new_student
```

```python
#score =  score.append(new_student)
score = pd.concat([score, new_student], ignore_index=False)
score
```

```python
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
new_student1
```

```python
score = pd.concat([score, new_student1.to_frame().T])
score
```

```python
# Dictionary 자료형을 사용하여 행을 추가할 경우 인덱스로 사용할 정보가 없는 경우 ignore_index인자를 True로 설정하지 않으면 오류가 발생
# ignore_index를 True로 설정하면 기존인덱스가 전부 사라지기 때문에 주의하여 사용하여야 한다.
# DataFrame에는 DataFrame 타입의 자료구조끼리만 합칠 수 있다.
new_student2 = {'국어':75, '수학':80, '영어':80}
score = pd.concat([score, new_student2], ignore_index = False)  # Error : 딕셔너리를 합치려고 하면 안됨
```

```python
# 임의의 조건 탐색 및 데이터 수정
import pandas as pd
students = pd.DataFrame({'이름':['장화','홍련','콩쥐','팥쥐','해님','달님'],
                        '국어':[70,85,None, 100, None, 85],
                        '수학':[65,100,80,95,None,70]})
students
```

```python
students[students['이름']=='장화']  # 조건이 하나인 경우,  =를 두번씀에 유의
```

```python
students[(students['국어'] >= 80) & (students['수학'] >= 80)]  # 조건이 두개인 경우
```

```python
students.loc[6, '이름':'수학'] = ['별님', 50, 60]  # 별님 추가
students
```

```python
students.loc[(students['국어']>=80)&(students['수학']>=70), "합격여부"] = 'Pass'  # 존재하지 않는 컬럼인 경우 새로 만듦
students.loc[students['합격여부']!='Pass', '합격여부'] = 'Fail'
students              
```

```python
# 하나의 컬럼에 여러 개의 조건에 따라 다른 값을 반환해야 하는 경우 numpy의 select() 이용
import numpy as np
condition_list = [(students['국어'] >= 90),
                  (students['국어'] >= 80) & (students['국어'] < 90),
                  (students['국어'] >= 70) & (students['국어'] < 80)]
choice_list = ['A', 'B', 'C']
students['점수'] = np.select(condition_list, choice_list, default='F')
students
```

```python
# 결측값 제거
'''
DataFrame.dropna(axis = 0, how = 'any', thresh = None, subset = None, inplace = False)
axis : 0은 행, 1은 열 삭제
how : 'any'이면 결측값이 존재하는 모든 행/열 삭제, 'all'이면 모든 값이 결측치일 때 삭제
thres : 정숫값을 지정하면 결측값이 아닌 값이 그보다 많을 때 행 또는 열을 유지
subset : 어떤 레이블에 결측값이 존재하면 삭제할지 정의
inplace : True이면 제자리에서 작업을 수행하고 None을 반환
'''
students.dropna()
```

```python
students.dropna(thresh=4)  # 결측값이 아닌 값이 4개 보다 많은 행만 남기기
```

```python
# 조인 : inner join(교집합), outer join(합집합), left join, right join 조인의 4가지가 있음. 
# join()과 merge() 두가지가 있고 merge()는 조금 더 다양한 조건을 사용할 수 있다.
'''
DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None,
                left_index=False, right_index=False, sort=False, suffixes=('_x','_y'))
right : 결합의 대상이 되는 프레임으로 right DataFrame
how : join의 방법
on : 키 칼럼
left_on/right_on : 두테이블에서 키 칼럼명이 다를 때 on을 대신하여 지정
sort : 키값을 기준으로 정렬 여부
suffixes : 결합한 데이터프레임에서 왼쪽/오른쪽 데이터프레임의 칼럼명에 사용할 접미사를 왼쪽, 오른쪽 순서로 지정
'''
product = pd.DataFrame({'상품코드':['G1','G2','G3','G4',],
                        '상품명':['우유','감자','빵','치킨',]})
product
```

```python
sales = pd.DataFrame({'주문번호':[1001, 1002, 1002, 1003, 1004],
                      '상품코드':['G4','G3','G1','G3','G5',],
                      '주문수량':[1, 4, 2, 2, 3,]})
sales
```

```python
sales.merge(product, on='상품코드', how='inner')
```

```python
sales.merge(product, on='상품코드', how='outer', sort=True)
```

```python
sales.merge(product, left_on='상품코드', right_on='상품코드', how='left')  # G2가 삭제되어 보임
```

```python
# 데이터 요약
'''
DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=False, dropna=True).FUN()
FUN() : 집계함수 ex) count(), sum(), min()...
'''
import pandas as pd
from sklearn.datasets import load_iris
IRIS = load_iris()
iris = pd.DataFrame(IRIS.data, columns=IRIS.feature_names)
IRIS
```

```python
iris['class'] = IRIS.target
iris
```

```python
iris['class'] = iris['class'].map({0:'setosa', 1:'versicolor', 2:'virginica'})  # 숫자를 문자로 바꾼다.
iris
```

```python
iris.groupby(by='class').mean()
```

```python
# 도수분포표
pd.Series(iris['class']).value_counts()  # 값의 갯수를 카운트
```

```python
iris['petal width level'] = pd.qcut(iris['petal width (cm)'], q=3, labels=['short','middle','long'])  # 구간에 따라 나눈다.
iris
```

```python
pd.Series(iris['petal width level']).value_counts().to_frame()
```

```python
pd.crosstab(iris['petal width level'], iris['class'])
```

```python
# 데이터프레임에 좀더 편하게 입력하는 방법, 딕셔너리 형태로 치는 것보다 쉬움
from io import StringIO
import pandas as pd

csv = """
학년,반,성별,성적
1, A, 여자, 76
1, A, 남자, 88
1, B, 여자, 85
1, B, 남자, 72
2, C, 여자, 68
2, C, 남자, 70
"""

score = pd.read_csv(StringIO(csv))
score
```

```python
# 피벗테이블
score = score.pivot_table(index=['학년','반'], columns='성별', values='성적')
score
```

```python
# 멜트 : 피벗테이블의 반대
score.reset_index()
```

```python
score.reset_index().melt(id_vars=['학년','반'], var_name='성별', value_name='성적')
```
