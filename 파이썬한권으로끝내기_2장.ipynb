```python
# 선택한 값 변경하는 법
score = pd.DataFrame({'국어':[100, 80], '수학':[75, 90],
                      '영어':[90,95]}, index=['장화', '홍련'])
score
```

```python
score.loc['홍련','영어'] = 100
score  # 영어성적이 100으로 바뀌었다.
```

```python
score['국어'] = score['국어'] - 5
score  # 모든 학생의 국어점수가 5점 내려감
```

```python
# 행과 열 추가
# 행추가는 append()를 사용
new_student = pd.DataFrame({'국어':[70,85], '수학':[65,100],
                            '영어':[95,65]}, index=['콩쥐', '팥쥐'])
new_student
```

```python
#score =  score.append(new_student)
score = pd.concat([score, new_student], ignore_index=False)
score
```

```python
new_student1 = pd.Series({'국어':85, '수학':55, '영어':95}, name='해님')
new_student1
```

```python
score = pd.concat([score, new_student1.to_frame().T])
score
```

```python
# Dictionary 자료형을 사용하여 행을 추가할 경우 인덱스로 사용할 정보가 없는 경우 ignore_index인자를 True로 설정하지 않으면 오류가 발생
# ignore_index를 True로 설정하면 기존인덱스가 전부 사라지기 때문에 주의하여 사용하여야 한다.
# DataFrame에는 DataFrame 타입의 자료구조끼리만 합칠 수 있다.
new_student2 = {'국어':75, '수학':80, '영어':80}
score = pd.concat([score, new_student2], ignore_index = False)  # Error : 딕셔너리를 합치려고 하면 안됨
```

```python
# DataFrame끼리만 연결되고 dict 타입은 연결이 안되므로 아래와 같이 한다.
new_student2_df = pd.DataFrame([new_student2], index=['달님'])
score = pd.concat([score, new_student2_df])
score
```

```python
# column 추가
science = [80, 70, 90, 85, 75, 70]
score['과학'] = science
score['학년'] = 1
score
```

```python
# 임의의 조건 탐색 및 데이터 수정
import pandas as pd
students = pd.DataFrame({'이름':['장화','홍련','콩쥐','팥쥐','해님','달님'],
                        '국어':[70,85,None, 100, None, 85],
                        '수학':[65,100,80,95,None,70]})
students
```

```python
students[students['이름']=='장화']  # 조건이 하나인 경우,  =를 두번씀에 유의
```

```python
students[(students['국어'] >= 80) & (students['수학'] >= 80)]  # 조건이 두개인 경우
```

```python
students.loc[6, '이름':'수학'] = ['별님', 50, 60]  # 별님 추가
students
```

```python
students.loc[(students['국어']>=80)&(students['수학']>=70), "합격여부"] = 'Pass'  # 존재하지 않는 컬럼인 경우 새로 만듦
students.loc[students['합격여부']!='Pass', '합격여부'] = 'Fail'
students              
```

```python
# 하나의 컬럼에 여러 개의 조건에 따라 다른 값을 반환해야 하는 경우 numpy의 select() 이용
import numpy as np
condition_list = [(students['국어'] >= 90),
                  (students['국어'] >= 80) & (students['국어'] < 90),
                  (students['국어'] >= 70) & (students['국어'] < 80)]
choice_list = ['A', 'B', 'C']
students['점수'] = np.select(condition_list, choice_list, default='F')
students
```

```python
# 결측값 탐색 및 수정
# 결측값 탐색함수 : isna(), isnull() (결측값일때)   notna(), notnull() (결측값이 아닌 값일때)
students.isna()
```

```python
students.notna()
```

```python
# 결측값 갯수 구하기
students.isna().sum()  # 열합계
```

```python
students.isna().sum(1)  # 행합계
```

```python
# 결측값 제거
'''
DataFrame.dropna(axis = 0, how = 'any', thresh = None, subset = None, inplace = False)
axis : 0은 행, 1은 열 삭제
how : 'any'이면 결측값이 존재하는 모든 행/열 삭제, 'all'이면 모든 값이 결측치일 때 삭제
thres : 정숫값을 지정하면 결측값이 아닌 값이 그보다 많을 때 행 또는 열을 유지
subset : 어떤 레이블에 결측값이 존재하면 삭제할지 정의
inplace : True이면 제자리에서 작업을 수행하고 None을 반환
'''
students.dropna()
```

```python
students.dropna(thresh=4)  # 결측값이 아닌 값이 4개 보다 많은 행만 남기기
```

```python
# 결측값 대체
'''
DataFrame.fillna(value = None, method=None, axis = None, inplace = False, limit = None)
value : 단일값 혹은 dict/Series/DataFrame 형식으로 대체할 값을 직접 입력
method : 'pad', 'ffill' 이전값으로 채우고, 'backfill', 'bfill'은 다음에 오는 값으로 채움
axis : 0은 행, 1은 열방향으로 채움
limit : method 인자를 지정한 경우 limit으로 지정한 갯수만큼 대체할 수 있음
'''
health = pd.DataFrame({'연도':[2017, 2018, 2019, 2020, 2021, 2022],
                       '키':[160, 162, 165, None, None, 166],
                       '몸무게':[53, 52, None, 50, 51, 54],
                       '시력':[1.2, None, 1.2, 1.2, 1.1, 0.8],
                       '병결':[None, None, None, 2, None, 1]})
health
```

```python
health.fillna(0)  # 결측치를 0으로 채움
```

```python
health.fillna(health.mean())  # 평균으로 결측값을 대체
```

```python
# 특정 칼럼의 결측치 처리
health['병결'] = health['병결'].fillna(0)   # 병결의 결측치는 0을 의미하는 경우
health
```

```python
health['몸무게'] = health['몸무게'].fillna(health['몸무게'].mean())  # 몸무게만 결측치를 평균으로 대체
health
```

```python
health.ffill(inplace=True)  # 키와 시력은 바로 직전값으로 대체
health
```

```python
# 중복행 삭제
'''
DataFrame.drop_duplicates()
'''
health['키'].drop_duplicates()  # 키의 중복을 거거
```
