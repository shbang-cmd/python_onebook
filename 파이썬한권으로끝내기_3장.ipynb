```python
# 파이썬 한 권으로 끝내기 3장
# EDA(Explatory Data Analysis)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
wine_load = load_wine()
wine = pd.DataFrame(wine_load.data, columns = wine_load.feature_names)
wine
```

```python
wine['Class'] = wine_load.target
wine
```

```python
wine['Class'] = wine['Class'].map({0:'class_0', 1:'class_1', 2:'class_2'})
wine
```

```python
# 도수분포표
wine_type = wine['Class'].value_counts()
wine_type
```

```python
# 막대그래프
plt.bar(wine_type.index, wine_type.values)
plt.show()
```

```python
# 세부 인자를 주어 보았다.
plt.bar(wine_type.index, wine_type.values, width=0.8, bottom=None, align='center')
plt.show()
```

```python
# barh() 함수 : 수평막대그래프
plt.barh(wine_type.index, wine_type.values)
plt.show()
```

```python
# 세부 인자를 주어 보았다.
plt.barh(wine_type.index, wine_type.values, height=0.8, left=None, align='edge')
plt.show()
```

```python
# 히스토그램 : 연속형 자료에 대한 도수분포표를 시각화
'''
plt.hist('변수명', bins=None, range=None, density=False, data=df)
bins : 히스토그램의 구간의 개수
range : bin의 상한갑과 하한값
density : True이면 확률밀도함수를 그리고 반환
'''
plt.title("Wine alcohol histogram")
plt.hist('alcohol', data=wine)
plt.show()
```

```python
plt.title("Wine alcohol histogram")
plt.hist('alcohol', bins=8, range=(11,15), color='purple', data=wine)  # 인자를 다양하게 주어보았다.
plt.show()
```

```python
# 상자그림(Box plot) : 사분위수를이용하여 수치형 변수값의 분포를 확인, IQR : Inter Quatile Range 중앙 50% 데이터들이 퍼진 정도
# 수염(whisker) : 1.5*IQ의 범위인 Q1 - 1.5*IQR 부터 Q1까지 Q3 + 1.5*IQR 부터 Q1까지 범위를 수염으로 지정하고 이 범위를 벗어나면 이상치로 표현
# IQR을 이상치로 정제하기 위한 판단기준으로 활용

import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1: 'Versicolour', 2:'Virginica'})
```

```python
# 하나의 컬럼만 boxplot으로 표현 가능
plt.boxplot(iris['sepal width (cm)'], whis=1.5) 
plt.title('sepal width (cm)')
plt.show()
```

```python
# 한 칼럼을 카테고리 변수에 따라 분리하여 boxplot으로 표현 가능
iris[['sepal width (cm)', 'class']].boxplot(by='class')
plt.show()
```

```python
# seaborn을 이용하여 더욱 시각화 가능
import seaborn as sns
sns.boxplot(x='class', y='sepal width (cm)', data=iris)
plt.show()
```

```python
# 산점도(scatter plot) : 두 개의 수치형 변수 각각의 분포와 함께 두 변수의 관계를 확인

import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1: 'Versicolour', 2:'Virginica'})

# 산점도 그리기 : Iris 데이터에서 꽃받침(sepal)의 길이 vs 너비 산점도를 그리는 코드
plt.title('iris scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(x=iris['sepal length (cm)'], y=iris['sepal width (cm)'], alpha=0.5) # alpha : 점의 투명도
plt.show
```

```python
# 한 칼럼을 카테고리 변수에 따라 색을 분리하여 scatterplot으로 표현

import seaborn as sns
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris, hue='class', style='class')
'''
x='sepal length (cm)' : x축에 꽃받침 길이
y='sepal width (cm)' : y축에 꽃받침 너비
data=iris : 데이터는 아까 만든 iris 데이터프레임 사용
hue='class' : 품종(class)에 따라 색깔을 다르게
style='class' : 품종(class)에 따라 마커 모양도 다르게
'''
plt.show()
```

```python
# 함수식 그래프
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1: 'Versicolour', 2:'Virginica'})

# 그래프
def linear_func(x):
    return 2*x + 1

X =  iris['sepal length (cm)']
plt.plot(X, linear_func(X), c='#789395')
plt.show()
         
```

```python
# 회귀선 그래프
'''
numpy.polyfit(X, Y, 차수)
'''
import numpy as np
X, Y = iris['sepal length (cm)'], iris['petal length (cm)']
b1, b0 = np.polyfit(X, Y, 1)  # 1차식
plt.scatter(x=X, y=Y, alpha=0.5)
plt.plot(X, b1*X+b0, color='red')
plt.show()
```

```python
# 2차식 그래프 : 1차식과 비슷하나 2차 이상의 그래프를 그리는 경우는 데이터를 X값에 대하여 정렬
iris2 = iris.sort_values(by='sepal length (cm)')
iris2
```

```python
X, Y = iris2['sepal length (cm)'], iris2['petal length (cm)']
b2, b1, b0 = np.polyfit(X, Y, 2)
plt.scatter(x=X, y=Y, alpha=0.5)
plt.plot(X, b0+b1*X + b2*X**2, color='red')
plt.show()
```

```python
# 꺽은선 그래프
plt.plot('sepal length (cm)', 'petal length (cm)', data=iris2)
plt.show()
```

```python
# 꺽은선 그래프도 카테고리별로 가능
plt.plot('sepal length (cm)', 'petal length (cm)', data = iris2.loc[iris2['class'] == 'Setosa'])
plt.plot('sepal length (cm)', 'petal length (cm)', data = iris2.loc[iris2['class'] == 'Versicolour'])
plt.plot('sepal length (cm)', 'petal length (cm)', data = iris2.loc[iris2['class'] == 'Virginica'])
plt.legend(iris2['class'].unique())
plt.show()
```

```python
# 상관관계 시각화
'''
scatter_matrix(data, alpha=0.5, figsize=(8,8), diagonal='hist')
data : 데이터프레임
alpha : 투명도(0~1)
figsize : 그래프 크기
diagonal : 대각선의 밀도 그래프 종료 {hist/kde}
'''
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1: 'Versicolour', 2:'Virginica'})

# 산점도 행렬
scatter_matrix(iris, alpha=0.5, figsize=(8,8), diagonal='hist')
plt.show()
```

```python
# 범주별 색깔을 가르게 나타낼 수 있다.
sns.pairplot(iris, diag_kind='auto', hue='class')
plt.show()
```

```python
# 상관계수 행렬 그래프
iris_corr = iris.drop(columns='class').corr(method='pearson')
sns.heatmap(iris_corr, xticklabels=iris_corr.columns, yticklabels=iris_corr.columns, cmap='RdBu_r', annot=True)
plt.show()
# 해석 : petal length와 petal width는 0.96으로 매우 강한 양의 상관관계를 가짐. 하지만 sepal length와 sepalwidth는 거의 상관이 없음.
```
