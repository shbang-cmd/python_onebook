```python
# 파이썬 한 권으로 끝내기 3장
# EDA(Explatory Data Analysis)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
wine_load = load_wine()
wine = pd.DataFrame(wine_load.data, columns = wine_load.feature_names)
wine
```

```python
wine['Class'] = wine_load.target
wine
```

```python
wine['Class'] = wine['Class'].map({0:'class_0', 1:'class_1', 2:'class_2'})
wine
```

```python
# 도수분포표
wine_type = wine['Class'].value_counts()
wine_type
```

```python
# 막대그래프
plt.bar(wine_type.index, wine_type.values)
plt.show()
```

```python
# 세부 인자를 주어 보았다.
plt.bar(wine_type.index, wine_type.values, width=0.8, bottom=None, align='center')
plt.show()
```

```python
# barh() 함수 : 수평막대그래프
plt.barh(wine_type.index, wine_type.values)
plt.show()
```

```python
# 세부 인자를 주어 보았다.
plt.barh(wine_type.index, wine_type.values, height=0.8, left=None, align='edge')
plt.show()
```
